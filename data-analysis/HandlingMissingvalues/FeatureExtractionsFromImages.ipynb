```python
# image processing and transformation
import skimage
import cv2
import numpy as np

from PIL import Image, ImageFilter
from sklearn.feature_extraction import image

import matplotlib.pyplot as plt
```

```python
img_stork = Image.open('bird.jpg')
```

```python
plt.figure(figsize=(12,6))

plt.imshow(img_stork)
plt.title('Original image')
plt.show()
```

```python
# access the size, format and mode of the image
img_stork.format, img_stork.mode, img_stork.size
```

```python
# convert into grayscale it's easier for ML to work with grayscale image data
gs_image = img_stork.convert(mode='L')

gs_image
```

```python
# check the size
gs_image.size
```

```python
# generate a smaller size of the original image
gs_image.thumbnail((200,200))

gs_image
```

```python
# check the size of original and croped image

gs_image.size , img_stork.size
```

```python
img_resize = img_stork.resize((300,300))

img_resize
```

```python
img_resize.size
```

```python
# flip the image by transpose, best practice to distort the original image
hoz_flip = img_stork.transpose(Image.FLIP_LEFT_RIGHT)
ver_flip = img_stork.transpose(Image.FLIP_TOP_BOTTOM)
```

```python
# plot the flipped image
plt.figure(figsize=(12, 6))

plt.title('Horizontal flip')
plt.imshow(hoz_flip)
plt.show()
```

```python
# plot the vertically flipped image
plt.figure(figsize=(12, 6))

plt.title('vertically flip')
plt.imshow(ver_flip)
plt.show()
```

```python
# images in form of array
img_stork_arr = np.array(img_stork)

type(img_stork_arr)
```

```python
img_stork_arr.shape
```

```python
# visualize the array representation of the image
plt.figure(figsize=(6,6))
plt.title('Numpy Image')
plt.imshow(img_stork_arr, cmap= 'gray')
```
