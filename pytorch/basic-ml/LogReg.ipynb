```python
# Logistics Regression
# Note: the output of LogReg is class base prediction
```

```python
# Import packages
import torch
from torch.autograd import Variable
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as f
```

```python
x = Variable(torch.Tensor([[25],[35],[45],[15]]))
y = Variable(torch.Tensor([[0],[1],[1],[0]]))
```

```python
# define the Model
class LogisticRegressionModel(nn.Module):
    def __init__(self, input_size, output_size):
        super(LogisticRegressionModel, self).__init__()
        self.linear = nn.Linear(input_size, output_size)
    
    def forward(self, x):
        y_predict = f.sigmoid(self.linear(x)) # to convert to 1 or 0 
        return y_predict
    
model = LogisticRegressionModel(1,1)
criteria = nn.BCELoss()
optimizer = optim.SGD(model.parameters(), 0.01)

for epoch in range(500):
    y_predict = model(x)
    loss =criteria(y_predict, y)
    optimizer.zero_grad() # clear the gradients.
    loss.backward() # calculate the back prop
    optimizer.step() # update the weights
    print(epoch, float(loss.data))
    

    test = Variable(torch.Tensor([[20]]))
    z = model.forward(test)
    print(float(z))
```
