```python
df = pd.DataFrame(data= {
    'feature_1': [np.nan,3,6,9,12,15,np.nan],
    'feature_2': [100, np.nan, 200, 300, np.nan, np.nan, 600],
    'feature_3': [1000, 500, 2000, 3000, 4000, 6000, 8000],
})

df
```

```python
 # return boolean of missing values
df.isnull()
```

```python
# check for total number of missing value
df.isnull().sum()
```

```python
# filling missing values
df.fillna(method='pad', limit=1)
```

```python
df.fillna(method='pad', limit=2)
```

```python
# back fill techniques for filling order data
df.fillna(method='bfill')
```

```python
# forward fill techniques for filling order data
df.fillna(method='ffill')
```

```python
# drop all missing data in the row in case of sufficient data 
df.dropna(axis=0) 
# (0=row, 1=columns)
```

```python
# drop all missing data in the column in case of sufficient data
df.dropna(axis=1)
```

```python
# drop a specific number of columns base on the quantity of missing data say above 10% in this case.
df.dropna(thresh=int(df.shape[0]* 0.9), axis=1)
```

```python
# filling all the missing values in the featur with the mean of that column
df['feature_1'].fillna(df['feature_1'].mean())
```
